##Ordering the data like first show female above green then male blue.

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
kashti=sns.load_dataset("titanic")

#draw bar plot
sns.barplot(x="sex",y="alone", hue="who", data=kashti, order=["female", "male"]) 
plt.show()

#Order command is used for arrangement.
```

```python

#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
kashti=sns.load_dataset("titanic")

#draw bar plot
sns.barplot(x="sex",y="alone", hue="who", data=kashti, order=["female", "male"], color="red") 
plt.show()
#color command used to change the color of plot. ex: "red","" (this called color palette)
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
kashti=sns.load_dataset("titanic")

#draw bar plot
sns.barplot(x="sex",y="alone", hue="who", data=kashti, order=["female", "male"], color="red", ci= None,
           palette="pastel") 
plt.show()
#palette command used for color change of bars.
```

##finding Mean, median using numpy library using import mean and import median

```python
#import libraries
import seaborn as sns
from numpy import mean 
#from numpy import median #for median import median then change estimator in sns.barplot(estimator=median)
import matplotlib.pyplot as plt

#Load dataset
kashti=sns.load_dataset("titanic")

#draw bar plot
sns.barplot(x="class", y="fare", hue="sex", data=kashti, estimator=mean)
#sns.barplot(x="class", y="fare", hue="sex", data=kashti, estimator=median)#estimator= median if import median as above discuss
plt.show()
#estimator command used for estimate like mean, median of bars.
```

##Color Saturation

```python
#import libraries
import seaborn as sns
from numpy import mean 
import matplotlib.pyplot as plt

#Load dataset
kashti=sns.load_dataset("titanic")

#draw bar plot
sns.barplot(x="class", y="fare", hue="sex", data=kashti, estimator=mean, saturation=0.2)
plt.show()
#saturation command used for color's saturation change of bars. 
```

```python
#import libraries
import seaborn as sns
from numpy import mean 
import matplotlib.pyplot as plt

#Load dataset
kashti=sns.load_dataset("titanic")

#draw bar plot
sns.barplot(x="fare", y="class", hue="sex", data=kashti, estimator=mean, saturation=1)
plt.show()
#saturation command used for color's saturation change of bars. 
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
kashti=sns.load_dataset("titanic")

#draw bar plot
sns.barplot(x="class", y="fare", hue="sex", data=kashti,
            linewidth=3, facecolor=(1, 0.3, 0.3, 0.2), 
           errcolor="0", edgecolor=".2")    
plt.show() 
#black is 0 error and white is 1 also in edge both same.#face color is color of buildings inside.
#linewidth is the building line width, errcolor is error bars, edge color is building width line color. 
```
