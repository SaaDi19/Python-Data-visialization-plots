# EDA on Spotify Dataset from the Kaggle

```python
#Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the dataset and checkinf first 5 data records.

```python
spotify_tracks=pd.read_csv("tracks.csv")
spotify_tracks.head()
```

### Checking the null values in dataset and total them

```python
pd.isnull(spotify_tracks).sum()
```

### total number of rows and columns and data types in dataset.

```python
spotify_tracks.info()
```

### Finding the 10 least popular songs

```python
sorted_spotify_tracks=spotify_tracks.sort_values('popularity',ascending=True).head(10)
sorted_spotify_tracks
```

```python
spotify_tracks.describe().transpose()
```

### finding the 10 most popuLAR SONGS

```python
most_popular=spotify_tracks.query('popularity>90',inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

### Changing the index of release date in dataset to datetime format

```python
spotify_tracks.set_index("release_date", inplace=True)
spotify_tracks.index=pd.to_datetime(spotify_tracks.index)
spotify_tracks.head()
```

### Checking artist at the 18th row by using index location method

```python
spotify_tracks[['artists']].iloc[18]
```

### Checking artist from the 18th row to 20 by using index location method

```python
spotify_tracks[['artists']].iloc[18:21]
```

### converting the duration in milisseconds to seconds and at new index after release date

```python
spotify_tracks["duration"]=spotify_tracks["duration_ms"].apply(lambda x:round(x/1000))
spotify_tracks.drop("duration_ms",inplace=True, axis=1)
```

```python
spotify_tracks.duration.head()
```

### Coreleationship heatmap of variables

```python
corr_spotify_tracks=spotify_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,16))
heatmap=sns.heatmap(corr_spotify_tracks, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0,cmap="inferno",linewidth=1,linecolor="Black")
plt.title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

### creating sample a data intack 0.4% of dataset

```python
sample_spotify_tracks=spotify_tracks.sample(int(0.004*len(spotify_tracks)))
```

```python
sample_spotify_tracks
```

### Correlation between loaudness and energy by regression line plot

```python
plt.figure(figsize=(12,5))
plt.title("Correleation between Loudness and Energey")
sns.regplot(x="loudness",y="energy", data=sample_spotify_tracks, color="c")
plt.xlim(-50)
plt.ylim(0)
```

### Correlation between popularity and acousticness by regression line plot

```python
plt.figure(figsize=(12,5))
plt.title("Correleation between Popularity and Acousticness")
sns.regplot(x="popularity",y="acousticness", data=sample_spotify_tracks, color="b")
#plt.xlim(-50)
#plt.ylim(0)
```

```python
spotify_tracks["dates"]=spotify_tracks.index.get_level_values("release_date")
spotify_tracks.dates=pd.to_datetime(spotify_tracks.dates)
years=spotify_tracks.dates.dt.year
```

```python
#Songs usability chart
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist")
plt.title("Number of songs per years history")
```

```python
#Durations of songs over the years using barplot
total_dur=spotify_tracks.duration
fig_dims=(18,7)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dur,ax=ax, errwidth=False)
plt.title("Year vs Duration")
plt.xticks(rotation=90)
```

```python
#Durations of songs over the years using  line plot
total_dur2=spotify_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig, ax=plt.subplots(figsize=fig_dims)
fig2=sns.lineplot(x=years,y=total_dur2,ax=ax)
plt.title("Durations of songs ")
plt.xticks(rotation=90)
```
