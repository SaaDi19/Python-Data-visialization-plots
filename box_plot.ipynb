```python
import seaborn as sns
import pandas as pd
import numpy as np

kashti=sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age", data=kashti)
```

```python

sns.boxplot(x="survived",
            y="age",
            showmeans=True,
            data=kashti)
sns.show()
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x="survived",
            y="age",
            showmeans=True,
            meanprops={"marker":"*","markersize":"12","markeredgecolor":"red"}, #marker=+ also acceptebale
            data=kashti)
#show labels
plt.xlabel("How many survived", size="14"),
plt.ylabel("Age (years)",size="12"),
plt.title("How many live or dead in titanic",size="16",weight='bold')
plt.show()
```
